## 2. Looking at the Mpro active site
----

<font size="3">Look at the following paper: </font>

```
Douangamath, A., Fearon, D., Gehrtz, P. et al. Crystallographic and electrophilic fragment screening of the SARS-CoV-2 main protease. Nat Commun 11, 5047 (2020). https://doi.org/10.1038/s41467-020-18709-w
```

<font size="3">
And pay particular attention to: the S labels given in reference to the active site, the term 'catalytic dyad'

Go back to fragalysis and complete the following exercises:

1. Turn on a surface for any ligand in the Aminopyridine site using the 'S' button on the ligands display controls

<img src="../img/ligand_controls.png">
    
2. Turn on all ligands in the aminopyridine site using the L button in the hit navigator menu bar
    
<img src="../img/hit_nav_menu.png">
    
3. Orient the central viewer so that you have a good view of the active site and ligands.
    **HINT**: you can see the mouse controls with the mouse icon below the cetral viewer. You can change basic display options using the cog icon below the central viewer
